16) Write a program to check if a year is leap year or not.

If a year is divisible by 4 then it is leap year but if the year is century year like 2000, 1900, 2100 then it must be divisible by 400.

```python
year = int(input("Enter a year: "))

if year % 100 == 0:

    if year % 400 == 0:
        print(year, "is a leap year")
    else:
        print(year, "is not a leap year") 
else:

    if year % 4 == 0:
        print(year,"is a leap year")
    else:
        print(year, "is not a leap year")            
```

```python
# Take number of units as input from the user
units = int(input("Enter the number of units: "))

# Initialize bill amount
bill = 0

# Calculate bill according to the criteria
if units <= 100:
    bill = 0
elif units <= 300:
    # First 100 units free, next units charged at Rs 5/unit
    bill = (units - 100) * 5
else:
    # First 100 units free, next 200 units at Rs 5/unit, remaining units at Rs 10/unit
    bill = (200 * 5) + (units - 300) * 10

# Display the total bill
print("Total electricity bill: Rs.", bill)
```
